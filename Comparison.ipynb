```python
import pandas as pd
import numpy as np
from sklearn import metrics
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# NB
fpr_nb = [0., 0.25808988, 1.]
tpr_nb = [0., 0.86314594, 1.]
roc_auc_nb = 0.80

# LR
fpr_lr = [0., 0.07028368, 1.]
tpr_lr = [0., 0.56407344, 1.]
roc_auc_lr = 0.75

#DT
fpr_dt = [0., 0.06880844, 1.]
tpr_dt = [0., 0.49883087, 1.]
roc_auc_dt= 0.72


# RF
fpr_rf = [0., 0.07346914, 1.]
tpr_rf = [0., 0.61064319, 1.]
roc_auc_rf = 0.77

# NN1
fpr_nn1 = [0., 0.04427509, 1.]
tpr_nn1 = [0., 0.38008707, 1.]
roc_auc_nn1 = 0.67

# NN2
fpr_nn2 = [0., 0.01457691, 1.]
tpr_nn2 = [0., 0.16505773, 1.]
roc_auc_nn2 = 0.58

# NN3
fpr_nn3 = [0., 0.28409847, 1.]
tpr_nn3 = [0., 0.8720424, 1.]
roc_auc_nn3 = 0.79

plt.title('Receiver Operating Characteristic')

plt.plot(fpr_lr, tpr_lr, 'b', label='Logistic Regression (AUC = %0.2f)' % roc_auc_lr)
plt.plot(fpr_rf, tpr_rf, 'r', label='Random Forest (AUC = %0.2f)' % roc_auc_rf)
plt.plot(fpr_nb, tpr_nb, 'y', label='Naive Bayes (AUC = %0.2f)' % roc_auc_nb)
plt.plot(fpr_dt, tpr_dt, 'k', label='Decision Tree (AUC = %0.2f)' % roc_auc_dt)
plt.plot(fpr_nn1, tpr_nn1, 'g', label='Neural Network-1 hidden layer (AUC = %0.2f)' % roc_auc_nn1)
plt.plot(fpr_nn2, tpr_nn2, 'c', label='Neural Network-2 hidden layer (AUC = %0.2f)' % roc_auc_nn2)
plt.plot(fpr_nn3, tpr_nn3, 'm', label='Neural Network-3 hidden layer (AUC = %0.2f)' % roc_auc_nn3)

plt.legend(loc='lower right')
plt.plot([0, 1], [0, 1], 'r--')
```
